```python
#Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('Diwali Sales Data.csv',encoding ='unicode_escape')
```

```python
data=data.drop_duplicates()
```

```python
data.shape
```

```python
data.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
data.columns
```

```python
data.isnull().sum()
```

```python
data=data.dropna()
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data['Amount']=data['Amount'].astype('int')
```

```python
data['Amount'].dtype
```

```python
data.head()
```

```python
data=data.rename(columns={'Cust_name':'Customer_Name'})
```

```python
data1=pd.DataFrame(data)
```

```python
data1['Marital_Status']=data1['Marital_Status'].astype("str")

```

```python
data1['Marital_Status'].dtype
```

```python
data1['Marital_Status'] = data1['Marital_Status'].replace({"1": 'Married', "0": 'Unmarried'})
```

```python
data1['Gender'] = data1['Gender'].replace({"F": "Female", "M": "Male"})
```

```python
data1.head(2)
```

```python
data1.describe()
```

```python
data1[['Age','Orders','Amount']].describe()
```

```python
data1.head(3)
```

```python
#Total Customer by Gender
ax=sns.countplot(x='Gender',data=data1)
plt.title('Customer by Gender')

for bars in ax.containers:
    ax.bar_label(bars)
    
    
```

```python
#Total Sales Revenue by Gender
sales_gen = data1.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',
ascending=False)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Gender', y='Amount', data=sales_gen, s=100, color='black') 
plt.title('Total Sales Revenue by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Sales Amount in ten-millions')
plt.show()
```

```python
"""As we can see the from above graphs the females are participating and are more interested 
sales than men.Also the purchasing power of Female is higher than male as the total sales revenue 
generated from female is much higher than that of male."""
```

```python
 #Total count of customers in each age-group data 
cx=sns.countplot(x='Age Group',data=data1)
plt.title("Total count of customers by age group")
for bars in cx.containers:
    cx.bar_label(bars)
plt.show()
```

```python
#Total Revenue Generated by each Age-Group
total_revenue_age = data1.groupby(['Age Group'], as_index =False).sum('Amount').sort_values(by='Amount',
ascending = False)
dx=sns.barplot(x = 'Age Group', y = 'Amount', data = total_revenue_age)
plt.title("Total Revenue Generated by each Age-Group")
for bars in dx.containers:
    dx.bar_label(bars)
plt.show()
```

```python
#Total Customer Count Vs Age-Group Based on Gender
plt.figure(figsize=(8, 6))
cx=sns.countplot(x="Age Group",data=data1,hue='Gender')
plt.title("Total Customer Count Vs Age-Group Based on Gender")

for bars in cx.containers:
    cx.bar_label(bars)
plt.show()
```

```python
grouped_data=data1.groupby("Age Group")
```

```python
sum_of_revenue=grouped_data['Amount'].sum()
sum_of_revenue
```

```python
#Contribution to Revenue by Age-Group
plt.pie(grouped_data['Amount'].sum(),labels=grouped_data['Age Group'].unique(),autopct='%0.1f%%')
plt.title('Contribution to Revenue by Age-Group')
plt.show()
```

```python
#Total Revenue Generated by Each States
total_revenue_state = data1.groupby(['State'],as_index =False).sum('Amount').sort_values(by='Amount',
ascending = False)
# plot the data
plt.figure(figsize=(35,6))
st = sns.barplot(x = 'State', y = 'Amount', data = total_revenue_state)
plt.title("Total Revenue Generated by Each States")
plt.ylabel("Total revenue generated")
plt.xlabel("States")
for bar in  st.containers:
    st.bar_label(bars)
plt.show()
```

```python
#Total orders by each State
total_order_state = data1.groupby(['State'], as_index =False).sum('Orders').sort_values(by='Orders',
ascending = False).head(5)
# plot the data
plt.figure(figsize=(35,6))
states_plot = sns.barplot(x = 'State', y = 'Orders', data = total_revenue_state)
plt.title("Total orders by each State")
plt.xlabel("States")

for bars in states_plot.containers:
    states_plot.bar_label(bars)
plt.show()
```

```python
""""Insights From States We can observe the graphs showing 5 states of India with highest sales, 
these states are Uttar Pradesh, Maharashtra, Karnataka, Dehli and Mahya Pradesh."""
```

```python
grouped_state = data.groupby("State").sum("Amount")
grouped_state
```

```python
average_order_price = grouped_state["Amount"]/grouped_state["Orders"]
average_order_price
```

```python
#Average Order Price In each State
plt.figure(figsize=(15,6))
states_plot = sns.barplot(x = grouped_state.index, y = average_order_price, data = grouped_state)
plt.title("Average Order Price In each State")
plt.ylabel("Average Order Price")
plt.xlabel("States")
plt.xticks(rotation = 80)

for bars in states_plot.containers:
    states_plot.bar_label(bars)
plt.show()
```

```python
"""Although Maharasthra is leading state in generating more sales revenue but for per unit cost Dehli is 
at the top of list.This can be either due to the high average price charged in Dehli as compare to other
states or Dehli people are purchasing more luxurios items costing them high prices."""
```

```python
#Total Revenue Generated by Marital_Status
status = data1.groupby(['Marital_Status'],as_index = False).sum("Amount").sort_values(by= "Amount",
ascending = False)
# plot the data
sns.set(rc={'figure.figsize':(6,6)})
status_plot = sns.barplot(x= 'Marital_Status', y = 'Amount', data = status)
plt.title("Total Revenue Generated by Marital_Status")
plt.ylabel("Revenue generated in 10 millions")

for bars in status_plot.containers:
    status_plot.bar_label(bars)
plt.show()
```

```python
#Total Revenue Generated by Marital_Status based on gender
status = data1.groupby(['Marital_Status', 'Gender'],as_index = False).sum("Amount").sort_values(by=
"Amount",ascending = False)
# plot the data
sns.set(rc={'figure.figsize':(6,6)})
status_plot = sns.barplot(x= 'Marital_Status', y = 'Amount', data = status, hue ='Gender')
plt.title("Total Revenue Generated by Marital_Status")
plt.ylabel("Revenue generated in 10 millions")
for bars in status_plot.containers:
    status_plot.bar_label(bars)
plt.show()    
```

```python
"""From the above graph and we can analyze that the most of sales are coming from Unmarried females and 
then second highest contribution to the sale is Married female."""
```

```python
#Total Customer Count by Occupation
occupation = data1.groupby(['Occupation'], as_index = False).sum("Amount").sort_values(by = 'Amount',
ascending = False)
sns.set(rc={'figure.figsize':(25,6)})
occupation_plot = sns.countplot(data = data1, x = 'Occupation')
plt.title("Total Customer Count by Occupation")
for bars in occupation_plot.containers:
    occupation_plot.bar_label(bars)
plt.show()
```

```python
#Top 5 Revenue Generated  Vs Occupation of customers
occupation_data = data1.groupby(['Occupation', 'Gender'], as_index = False).sum("Amount").sort_values(by=
'Amount', ascending = False).head(5)
# plot the data
status_plot = sns.barplot(x= 'Occupation', y = 'Amount', data = occupation_data)
plt.title("Top 5 Revenue Generated  Vs Occupation of customers")
```

```python
"""As we observe the graph we noticed that the IT Sector generats the highest sales revenue and 
after that Aviation,Healthcare and then Banking and so on."""
```

```python
#Products have highest orders
sns.set(rc={'figure.figsize': (20,5)})
sns.countplot(x = 'Product_Category', data = data1)
plt.xlabel("Product Category")
plt.ylabel("Number of orders")
plt.show()
```

```python
#TOTAL REVENUE GENERATED BY TOP 5 PRODUCTS
product = data1.groupby(['Product_Category'], as_index=False).sum('Amount').sort_values(by = 
'Amount',ascending = False).head(5)
#plot the data
sns.set(rc={'figure.figsize':(10,6)})
product_plot = sns.barplot(x = 'Product_Category', y = 'Amount', data = product)
plt.title("TOTAL REVENUE GENERATED BY TOP 5 PRODUCTS")
plt.ylabel("Amount in 10 millions")
for bars in product_plot.containers:
    product_plot.bar_label(bars)
plt.show()    
```

```python
"""We can see a little deflection in data for the product category. For clothing & Apparel 
the number of order is the highest but for revenue generation Food is the highest contributor."""
```
